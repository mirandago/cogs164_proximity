Commute Challenges

```python
#"commute_challenges_mwf_on_campus", "commute_challenges_tth_on_campus", "commute_challenges_ss_on_campus"
def clean_challenges (row):
    cleaned = ''
    for entry in [row["commute_challenges_mwf_off_campus"], row["commute_challenges_tth_off_campus"],
                  row["commute_challenges_ss_off_campus"]]:
        print(entry)
        if entry is np.nan:
            cleaned += 'X'
            continue
        if 'uphill' in entry:
            cleaned += 'U'
        if 'stair' in entry:
            cleaned += 'S'
    return cleaned
```

```python
challenge['week_challenge'] = challenge.apply(lambda row: clean_challenges(row), axis=1)
```

```python
none = challenge[challenge['week_challenge'].str.contains("XXX")]
uphill = challenge[challenge['week_challenge'].str.contains("U")]
stairs = challenge[challenge['week_challenge'].str.contains("S")]
```

```python
none_freq = none['off_campus_gym_freq'].mean()
uphill_freq = uphill['off_campus_gym_freq'].mean()
stairs_freq = stairs['off_campus_gym_freq'].mean()
```

```python
objects = ('None', 'Uphill', 'Stairs')
y_pos = np.arange(len(objects))
challenge_off_freq = [none_freq,uphill_freq, stairs_freq]

plt.bar(y_pos, challenge_off_freq, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Count of Off Campus Gym Frequencies')
plt.title('Mean of Off Campus Gym Frequencies based on Challenges')

plt.show()
```

Transportation

```python
def clean_transport (row):
    cleaned = ''
    for entry in [row["transportation_mwf_on_campus"], row["transportation_tth_on_campus"],
                  row["transportation_ss_on_campus"]]:
        if entry is np.nan:
            cleaned += 'X'
            continue
        if 'Walk' in entry:
            cleaned += 'W'
        if 'Bike' in entry:
            cleaned += 'B'
        if 'Car' in entry:
            cleaned += 'C'
        if 'Skateboard' in entry:
            cleaned += 'K'
        if 'Bus' in entry:
            cleaned += 'S'
    return cleaned
```

```python
transport['week_transport'] = transport.apply(lambda row: clean_transport(row), axis=1)
```

```python
none = transport[transport['week_transport'].str.contains('XXX')]
walk = transport[transport['week_transport'].str.contains("W")]
bike = transport[transport['week_transport'].str.contains("B")]
car = transport[transport['week_transport'].str.contains("C")]
skateboard = transport[transport['week_transport'].str.contains("K")]
shuttle = transport[transport['week_transport'].str.contains("S")]
```

```python
none_freq = none['off_campus_gym_freq'].mean()
walk_freq = walk['off_campus_gym_freq'].mean()
bike_freq = bike['off_campus_gym_freq'].mean()
car_freq = car['off_campus_gym_freq'].mean()
skateboard_freq = skateboard['off_campus_gym_freq'].mean()
shuttle_freq = shuttle['off_campus_gym_freq'].mean()
walk_freq
```

```python
objects = ('None','Walk/Jog', 'Bike', 'Car', 'Skateboard', 'Shuttle')
y_pos = np.arange(len(objects))
transport_off_freq = [none_freq, walk_freq, bike_freq, car_freq, skateboard_freq, shuttle_freq]

plt.bar(y_pos, transport_off_freq, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Count of Off Campus Gym Frequencies')
plt.title('Mean of Off Campus Gym Frequencies based on Mode of Transportation')

plt.show()
```
